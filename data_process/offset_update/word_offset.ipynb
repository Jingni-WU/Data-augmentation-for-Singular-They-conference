```python
import pandas as pd 
import spacy 

file_path = 'fem_data.csv'
df = pd.read_csv(file_path, sep='\t')

```

```python
df.head()
```

```python
nlp = spacy.load("en_core_web_sm")

def find_word_offset(text, char_offset):
    # tokenizes text 
    doc = nlp(text)
    word_offset = 0
    char_count = 0
    
    for token in doc:
        # for every token, add to char_count (add length of the token's text + whitespace) 
        char_count += len(token.text_with_ws)
        if char_count > char_offset:
            break
        else:
            if not token.is_space:
                # calc word offset 
                word_offset += 1
    
    return word_offset

```

```python
new_pron_offsets = []

for index, row in df.iterrows():
    
    original_sentence = df.loc[index, 'Text']
    original_offset = df.loc[index, 'Pronoun-offset']

    word_offset = find_word_offset(original_sentence, original_offset)
    
    new_pron_offsets.append(word_offset)
    
    
df['NEW_pronoun_offset'] = new_pron_offsets
```

```python
new_A_offsets = []

for index, row in df.iterrows():
    
    original_sentence = df.loc[index, 'Text'] 
    original_offset = df.loc[index, 'A-offset']

    word_offset = find_word_offset(original_sentence, original_offset)
    
    new_A_offsets.append(word_offset)
    
    
df['NEW_A-offset'] = new_A_offsets
```

```python
new_B_offsets = []

for index, row in df.iterrows():
    
    original_sentence = df.loc[index, 'Text'] 
    original_offset = df.loc[index, 'B-offset']

    word_offset = find_word_offset(original_sentence, original_offset)
    
    new_B_offsets.append(word_offset)
    
    
df['NEW_B-offset'] = new_B_offsets
```

```python
df.head()
```

```python
final_df = df[['ID', 'Text', 'Pronoun', 'NEW_pronoun_offset', 'A', 'NEW_A-offset', 'A-coref', 'B', 'NEW_B-offset', 'B-coref', 'URL']]
```

```python
new_column_names = {'NEW_pronoun_offset': 'Pronoun-offset', 'NEW_A-offset': 'A-offset', 'NEW_B-offset': 'B-offset'}
final_df.rename(columns=new_column_names, inplace=True)
```
