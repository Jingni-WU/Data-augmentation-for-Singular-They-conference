```python
import spacy

nlp = spacy.load("en_core_web_sm")

doc = nlp(test_sentence)

# First loop to find verbs connected with 'they'
verbs_with_they = []
for token in doc:
    # check if token is a verb of 'they'
    if token.pos_ == "VERB" and "they" in [child.text for child in token.children]:
        verbs_with_they.append(token)

# Second loop to find aux connected with those verbs
auxiliary_verbs = []
# check if token is an aux of verb
for verb_token in verbs_with_they:
    for child in verb_token.children:
        if child.pos_ == "AUX":
            auxiliary_verbs.append(child.text)

print(verbs_with_they)
print("Auxiliary verbs connected to the verbs:", auxiliary_verbs)
```
