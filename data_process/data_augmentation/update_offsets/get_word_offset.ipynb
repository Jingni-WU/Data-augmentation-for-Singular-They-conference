```python
nlp = spacy.load("en_core_web_sm")

def find_word_offset(text, char_offset):
    doc = nlp(text)
    word_offset = 0
    char_count = 0
    
    for token in doc:
        char_count += len(token.text_with_ws)
        if char_count > char_offset:
            break
        else:
            if not token.is_space:
                word_offset += 1
    
    return word_offset

```

```python
new_pron_offsets = []

for index, row in df.iterrows():
    
    original_sentence = df.loc[index, 'Text']
    original_offset = df.loc[index, 'Pronoun-offset']

    word_offset = find_word_offset(original_sentence, original_offset)
    
    new_pron_offsets.append(word_offset)
    
    
df['NEW_pronoun_offset'] = new_pron_offsets
```

```python
new_A_offsets = []

for index, row in df.iterrows():
    
    original_sentence = df.loc[index, 'Text'] 
    original_offset = df.loc[index, 'A-offset']

    word_offset = find_word_offset(original_sentence, original_offset)
    
    new_A_offsets.append(word_offset)
    
    
df['NEW_A-offset'] = new_A_offsets
```

```python
new_B_offsets = []

for index, row in df.iterrows():
    
    original_sentence = df.loc[index, 'Text'] 
    original_offset = df.loc[index, 'B-offset']

    word_offset = find_word_offset(original_sentence, original_offset)
    
    new_B_offsets.append(word_offset)
    
    
df['NEW_B-offset'] = new_B_offsets
```

```python
nlp = spacy.load("en_core_web_sm")

def replace_word_at_index(text, index, replacement):

    doc = nlp(text)
    words = [token.text for token in doc]
    
    # Replace the word at the specified index
    if 0 <= index < len(words):
        words[index] = replacement
    
    # Reconstruct the modified string
    modified_text = ' '.join(words)
    
    return modified_text
```

```python


FINAL_TEXT = []

for index, row in df.iterrows():
    
    original_sentence = str(df.loc[index, 'Text'])
    altered_sentence = str(df.loc[index, 'FINAL_inflected_text'])
    original_offset = int(df.loc[index, 'Pronoun-offset'])
    target_word = str(df.loc[index, 'New_Pronoun'])

    word_offset = find_word_offset(original_sentence, original_offset)  
    
    mod_text = replace_word_at_index(altered_sentence, word_offset, target_word)

    
    FINAL_TEXT.append(mod_text)
```

```python
df['FINAL_TEXT'] = FINAL_TEXT
```
