```python
import pandas as pd 
df = pd.read_csv('10000.csv') 
df.head()
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

def word_offsets_to_char_offsets(text, word_offsets):
    doc = nlp(text)
    char_offsets = []
    current_offset = 0

    for token in doc:
        if token.i in word_offsets:
            start_char_offset = current_offset
            end_char_offset = current_offset + len(token.text)
            char_offsets.append((start_char_offset, end_char_offset))

        current_offset += len(token.text_with_ws)

    return char_offsets


```

```python
new_A_offsets_start = []
new_A_offsets_end = []

for index, row in df.iterrows():
    
    text = df.loc[index, 'Text']
    word_offsets = [df.loc[index, 'A-offset']]

    char_offsets = word_offsets_to_char_offsets(text, word_offsets)
    
    for offset in char_offsets:
        start_offset = offset[0]
        end_offset = offset[1]
        new_A_offsets_start.append(start_offset)
        new_A_offsets_end.append(end_offset)
        
df['A_offset_start'] = new_A_offsets_start
    
    
df['A_offset_end'] = new_A_offsets_end
```

```python
new_B_offsets_start = []
new_B_offsets_end = []

for index, row in df.iterrows():
    
    text = df.loc[index, 'Text']
    word_offsets = [df.loc[index, 'B-offset']]

    char_offsets = word_offsets_to_char_offsets(text, word_offsets)
    
    for offset in char_offsets:
        start_offset = offset[0]
        end_offset = offset[1]
        new_B_offsets_start.append(start_offset)
        new_B_offsets_end.append(end_offset)
        
df['B_offset_start'] = new_B_offsets_start
    
    
df['B_offset_end'] = new_B_offsets_end
```

```python
new_pron_offsets_start = []
new_pron_offsets_end = []

for index, row in df.iterrows():
    
    text = df.loc[index, 'Text']
    word_offsets = [df.loc[index, 'Pronoun-offset']]

    char_offsets = word_offsets_to_char_offsets(text, word_offsets)
    
    for offset in char_offsets:
        start_offset = offset[0]
        end_offset = offset[1]
        new_pron_offsets_start.append(start_offset)
        new_pron_offsets_end.append(end_offset)
        
df['PRON_offset_start'] = new_pron_offsets_start
    
    
df['PRON_offset_end'] = new_pron_offsets_end
```
