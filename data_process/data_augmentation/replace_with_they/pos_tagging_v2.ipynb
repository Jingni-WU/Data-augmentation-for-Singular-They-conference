```python
import pandas as pd
import stanza
import spacy

# download stanza model
nlp = stanza.Pipeline(lang='en', processors="tokenize,pos", package={"tokenize": "spacy"})

# for using an example sentence from GAP data
facts = pd.read_csv('all_gap_data.tsv', delimiter='\t', encoding='utf-8')
df = facts.copy()

def get_pos_tags(sentence):
    doc = nlp(sentence)
    # sent_list = [sent.text for sent in doc.sentences]
    pos_tags = [f'{word.text}_{word.xpos}' for sent in doc.sentences for word in sent.words]
    return pos_tags

df['POS_Tags'] = df['Text'].apply(get_pos_tags)
df
```

```python
df.to_csv('pos_tagged_all_gap_v2.csv',index=False)
print("File exported.")
```
