```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('updated_pos_mapping_gender_neutral_data.csv', encoding='utf-8')
print(df.shape)
df.head()
```

EXAMPLES: 

He/she (PRP) → they 
She went to the park.
They went to the park.

his/her (PRP$)  → their
Alex needed to tie her shoe.
Alex needed to tie their shoe.

him/her  (PRP) → them
I didn’t see him at the party.
I didn’t see them at the party.

his/hers (PRP$) → theirs
The book left in class was his.
The book left in class was theirs.

```python
# Create new column called "New_Pronoun" with the correct "they" form based on POS tag of original gendered pronoun


# GENDER GROUPS (upper and lower case)
# she_PRP / he_PRP --> they
prp_nom = ["she_PRP", "he_PRP"]
PRP_nom = ["She_PRP","He_PRP"]

# her_PRP / him_PRP --> them
prp_obj = ["her_PRP","him_PRP"]
PRP_obj = ["Her_PRP","Him_PRP"]

# her_PRP$ / his_PRP$ --> their
prps_poss = ["his_PRP$","her_PRP$"]
PRPS_poss = ["His_PRP$", "Her_PRP$" ]

# EXCLUDING BC IT DOESN'T EXIST IN THE DATA AND MAY CAUSE ERRORS IN FUTURE TASKS FOR NOW
# hers_PRP$ / his_PRP$ --> theirs
# prps_plural = ["his","hers"]

# set values in new column to correct they form (w/ casing) based on group
df["New_Pronoun"] = np.select([df["pron_pos_tag"].isin(prp_nom), df["pron_pos_tag"].isin(PRP_nom), 
                                df["pron_pos_tag"].isin(prp_obj), df["pron_pos_tag"].isin(PRP_obj),
                                df["pron_pos_tag"].isin(prps_poss),df["pron_pos_tag"].isin(PRPS_poss)],
                            ["they", "They", "them","Them","their","Their"],default=None)

df.head()
```

```python
df.to_csv('gender_neutral_data_verb_step1.csv')
```
