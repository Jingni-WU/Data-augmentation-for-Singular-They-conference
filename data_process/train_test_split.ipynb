```python
import pandas as pd 
```

```python
file_path = 'FEM_balanced_data2.csv'
df_fem = pd.read_csv(file_path) 
```

```python
test_size = int(len(df_fem) * 0.2)
df_test_fem = df_fem.iloc[:test_size]
df_train_fem = df_fem.iloc[test_size:]


df_test_fem.reset_index(drop=True, inplace=True)
df_train_fem.reset_index(drop=True, inplace=True)

print(len(df_train_fem))
df_train_fem.head()
```

```python
file_path = 'NEUTRAL_balanced_data2.csv'
df_neutral = pd.read_csv(file_path) 

```

```python
test_size = int(len(df_neutral) * 0.2)
df_test_neu = df_neutral.iloc[:test_size]
df_train_neu = df_neutral.iloc[test_size:]


df_test_neu.reset_index(drop=True, inplace=True)
df_train_neu.reset_index(drop=True, inplace=True)

print(len(df_train_neu))
df_train_neu.head()
```

```python
file_path = 'MASC_balanced_data2.csv'
df_masc = pd.read_csv(file_path) 
```

```python
test_size = int(len(df_masc) * 0.2)
df_test_masc = df_masc.iloc[:test_size]
df_train_masc = df_masc.iloc[test_size:]


df_test_masc.reset_index(drop=True, inplace=True)
df_train_masc.reset_index(drop=True, inplace=True)

print(len(df_train_masc))
df_train_masc.head()
```

```python
df_test = pd.concat([df_test_fem, df_test_neu, df_test_masc], ignore_index=True)
df_test.reset_index(drop=True, inplace=True)
test = df_test.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
test['ID'] = test['ID'].astype(str).str.split('-').str[1]
test['ID'] = 'test-' + test['ID']

print(len(test))
test.head()
```

```python
df_train = pd.concat([df_train_fem, df_train_neu, df_train_masc], ignore_index=True)
df_train.reset_index(drop=True, inplace=True)
train = df_train.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python

train['ID'] = train['ID'].astype(str).str.split('-').str[1]
train['ID'] = 'train-' + train['ID']

print(len(train))
train.head()
```
