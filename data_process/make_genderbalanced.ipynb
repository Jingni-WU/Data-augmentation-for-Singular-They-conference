```python
import pandas as pd

file_paths = ['gap-test.tsv', 'gap-train.tsv', 'gap-validation.tsv']

dfs = []

for file_path in file_paths:
    df = pd.read_csv(file_path, sep='\t')
    dfs.append(df)
    
for i, df in enumerate(dfs):
    # Find the number of rows in the DataFrame
    num_rows = len(df)
    print(f"Number of rows in DataFrame {i+1}: {num_rows}")
    
concatenated_df = pd.concat(dfs)
concatenated_df



```

```python
pronoun_counts = concatenated_df['Pronoun'].value_counts()
pronoun_counts
```

```python
fem_df = concatenated_df[concatenated_df['Pronoun'].isin(['she', 'her', 'She', 'Her', 'hers'])]
masc_df = concatenated_df[concatenated_df['Pronoun'].isin(['his', 'he', 'He', 'him', 'His'])]
```

```python
# fem data 
# Randomly shuffle rows
fem_df_shuffled = fem_df.sample(frac=1, random_state=42) 

# Get the split point
fem_split_idx_1 = int(1/3 * len(fem_df_shuffled))

# Split the shuffled df
fem_df_1_3 = fem_df_shuffled.iloc[:fem_split_idx_1]
new_fem_df = fem_df_shuffled.iloc[fem_split_idx_1:]

print("1/3 of the rows:")
print(len(fem_df_1_3))
print("\n2/3 of the rows:")
print(len(new_fem_df))
```

```python
# masc data 
# Randomly shuffle rows
masc_df_shuffled = masc_df.sample(frac=1, random_state=42) 

# Get the split point
masc_split_idx_1 = int(1/3 * len(masc_df_shuffled))

# Split the shuffled df
masc_df_1_3 = masc_df_shuffled.iloc[:masc_split_idx_1]
new_masc_df = masc_df_shuffled.iloc[masc_split_idx_1:]

print("1/3 of the rows:")
print(len(masc_df_1_3))
print("\n2/3 of the rows:")
print(len(new_masc_df))
```

```python
gender_neut_df = pd.concat([fem_df_1_3, masc_df_1_3], axis=0)
# shuffle 
gender_neutral_df = gender_neut_df.sample(frac=1, random_state=42) 
print(len(gender_neutral_df))
```
